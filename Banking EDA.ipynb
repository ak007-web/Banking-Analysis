**Creating bins for data**

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Med','High']

df['Income Band']=pd.cut(df['Estimated Income'], bins=bins, labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

**Categorical Columns**

```python
cat_columns= df[["GenderId","BRId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","IAId","Properties Owned","Risk Weighting","Income Band"]].columns

for col in cat_columns: 
    print(f"value counts for '{col}':")
    display(df[col].value_counts())
```

**Bar Graph**

```python
for i,predictor in enumerate(df[["GenderId","BRId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","IAId","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
for i,predictor in enumerate(df[["GenderId","BRId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","IAId","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

**Histplot**

```python
for col in cat_columns:
    if col=="Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of occupation count')
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

**Numerical Analysis**

```python
num_columns=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

plt.figure(figsize=(20,20))
for i,col in enumerate(num_columns):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show() 
```

**Heatmap**

```python
corr_matrix=df[num_columns].corr()
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='crest')
```
